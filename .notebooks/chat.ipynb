```python
from __future__ import annotations

from docling.document_converter import DocumentConverter

document_converter = DocumentConverter()

PAGE_BREAK = "---PAGE_BREAK---"
```

```python
doc = document_converter.convert("../data/Group_Project.pdf")
md = doc.document.export_to_markdown(page_break_placeholder=PAGE_BREAK)

md_pages = [page.strip() for page in md.split(PAGE_BREAK) if page.strip()]
md_pages[6]
```

```python
def get_page_content(path: str, page_number: int) -> str:
    doc = document_converter.convert(path, page_range=(page_number, page_number))
    return doc.document.export_to_markdown()
```

```python
page_content = get_page_content("../data/Group_Project.pdf", page_number=7)
page_content
```
